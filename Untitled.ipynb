```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from wordcloud import WordCloud
from textblob import TextBlob
```

```python
tweets=api.user_timeline(screen_name='sundarpichai',count=100,lang='en',tweet_mode='extended')
```

```python
i=1
for tweet in tweets[0:5]:
    print(str(i)+' - '+tweet.full_text)
    i+=1
    
```

# creating dataset

```python
dataset=pd.DataFrame([tweet.full_text for tweet in tweets],columns=['Tweets'])
```

```python
dataset
```

# data cleaning

```python
def clean_text(text):
    text = re.sub(r'@[A-Za-z]+','',text) #removed @mentions
    text = re.sub(r'#','',text) #Removing the # the hyper string symbol
    text = re.sub(r'RT[\s]','',text) #Removing RT
    text = re.sub(r'http?:\/\/S+','',text) # remove 
    text = re.sub(r'https:+','',text) # remove 
    text = re.sub(r':','',text) # remove : 
    return text
```

```python
dataset['Tweets']=dataset['Tweets'].apply(clean_text)
```

```python
dataset
```

```python
def get_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity
```

```python
def get_polarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
dataset['Subjectivity']=dataset['Tweets'].apply(get_subjectivity)
```

```python
dataset['Polarity']=dataset['Tweets'].apply(get_polarity)
```

```python
dataset
```

```python
all_words=" ".join([t2t for t2t in dataset['Tweets']])
```

```python
all_words
```

```python
word_cloud= WordCloud(height=500,width=700,random_state=26,max_font_size=120).generate(all_words)
```

```python
plt.imshow(word_cloud,interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
def get_analysis(score):
    if score<0:
        return 'Negative'
    elif score==0:
        return 'Neutral'
    else:
        return 'Positive'
    
```

```python
dataset['Analysis']=dataset['Polarity'].apply(get_analysis)
```

```python
dataset
```

```python
j=1
sort_df=dataset.sort_values(by=['Polarity'])
for i in range(0,sort_df.shape[0]):
    if sort_df['Analysis'][i]=='Positive':
        print(str(j)+" )"+sort_df['Tweets'][i])
        j+=1
```

```python
j=1
sort_df=dataset.sort_values(by=['Polarity'],ascending=False)
for i in range(0,sort_df.shape[0]):
    if sort_df['Analysis'][i]=='Negative':
        print(str(j)+" )"+sort_df['Tweets'][i])
        j+=1
```
